```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
# Calculate order value
order_value = order_items.groupby('order_id')['total_price'].sum().reset_index()
order_value.rename(columns={'total_price': 'order_value'}, inplace=True)
```

```python
# Merge
orders_merged = pd.merge(orders, order_value, on='order_id')
```

```python
# Reference date
reference_date = orders_merged['order_date'].max() + pd.Timedelta(days=1)
```

```python
# RFM Calculation
rfm = orders_merged.groupby('customer_id').agg({
    'order_date': lambda x: (reference_date - x.max()).days,
    'order_id': 'nunique',
    'order_value': 'sum'
}).reset_index()
```

```python
rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']
```

```python
rfm['segment'] = pd.cut(rfm['monetary'], bins=[0, 1000, 5000, rfm['monetary'].max()],labels=['Low', 'Medium', 'High'])
```
